Arthur Samuel, an American pioneer in the field of artificial intelligence, coined the term "**Machine Learning**" in 1959 while at IBM.

Evolved from the study of pattern recognition and computational learning theory in artificial intelligence, machine learning explores the study and construction of algorithms that can learn from and make predictions on data – such algorithms overcome following strictly static program instructions by making data-driven predictions or decisions, through building a model from sample inputs.

Samuel defines machine learning as the subfield of computer science that gives computers the ability to learn without being explicitly programmed.

# History of machine learning

Machine learning, today, is considered as one of the state of the art technologies and has been gathering a huge interest from academia,innovaters and entrepreneurs.Interest in machine learning is not always this smooth.It went through its cycles of ups and downs.Let us look at the breif history of machine learning since its dawn.

### 1950s - 1970s Budding of Ideas

###### 1950:
The first idea of machine learning is emanted from Alan Turing, the inventor of computer, in 1950 which is called turing's learning machine. Turing in his paper titled 'Computing Machinery and Intelligence' discussed about the prospects of creating an artificially intelligent machine. He argued that it might be possible to inject an idea into a machine and the machine performs the task to fall off into quiescence again.

###### 1952:
First Neural network machine, the SNARC, was developed in Marvin Minsky who is widely regarded as father of artificial intelligence. This machine is considered one of the first pioneering attempts at the field of artificial intelligence. Ironically he is also critised by many for the 'AI winter' that followed after publishing his book 'Perceptrons' in his late 70s

###### 1956:
In summer of of 1956 at Darthmouth college, ten scientists sharing an interest in neural nets,automata theory and the study of intelligence convened for a six week workshop.This Darthmouth Summer project is often regarded as the beginning of artificial intelligence as a field of research.

###### 1957:
In 1957, Frank Rosenblatt invents the perceptron.The invention of the perceptron generated a great deal of excitement and widely covered in the media.

###### 1967:
The nearest neighbor algorithm was created, which is the start of basic pattern recognition. The algorithm was used to map routes. 

## 1970s Gloomy AI Winter

The performance of all the early systems developed suffered because of poor methods for handling uncertainty, data scarcity, and severe hardware limitations on memory capacity and processor speed. 

Apart from these, a book authored by Minsky, Perceptrons, which became the foundational work in the analysis of limitations of artificial neural networks is also believed to have discouraged research of Neural networks. This book is the center of a controversy in the history of AI, as some claim it to have had great importance in the dawn of 'AI Winter' in the 1970s.

By the mid-1970s, there was a growing awareness of these problems. The realization that many AI projects could never make good on their initial promises led to the onset of the first '**AI winter**' a period of retrenchment, during which funding decreased and skepticism increased,and AI fell out of fashion.

## 1980s Turn of Tide

###### 1979:
Students at Stanford University develop a cart that can navigate and avoid obstacles in a room which came to be known as standford cart.

###### 1980:
Kunihiko Fukushima first publishes his work on the Neocognitron, a type of artificial neural network. Neocognition later inspired convolutional neural networks.

###### Early 1980s:
A new springtime arrived in the early 1980s, when Japan launched its FifthGeneration Computer Systems Project, a well-funded public–private partnership that aimed to leapfrog the state of the art by developing a massively parallel computing architecture that would serve as a platform for artificial intelligence. This occurred at peak fascination with the Japanese “post-war economic miracle,” a period when Western government and business leaders anxiously sought to divine the formula behind Japan’s economic success in hope of replicating the magic at home. When Japan decided to invest big in AI, several other countries followed suit.

###### Late 1980s:
Apart from the development of advanced Recurrent Neural networks and Reinforced Learning techniques, Commercialisation of machine learning packages also took place. This period also saw an increased participation not only from acedemia but also from various other fields of study.

# early 1990s Second AI winter :

Despite the break throughs in the earlier period, Second AI winter loomed. Few of the reasons are that contributed are failure of various governmental efforts to proliferate AI technologies, costs assosiated with employing expert systems for various businesses and inadequate computing,memory and data capacities.

This period too gradually rescinded in late 90s and is on high note since then.

# late 1990s Optimistic vibes:

###### 1992:
Gerald Tesauro develops TD-Gammon, a computer backgammon program that utilises an artificial neural network. TD-Gammon is able to rival, but not consistently surpass, the abilities of top human backgammon players. Though there are many previous game playing AIs, this marked the end of second AI winter.

###### 1995: 
Tin Kam Ho publishes a paper describing Random decision forests.

###### 1997:
IBM’s Deep Blue beats the world champion, kasperov at chess. This acheivement is significant because only a couple of decades earlier there were few eminent researchers who regarded that playing a chess game is a human thing and it would take not less than a century to achieve that.

###### 2006: 
The Netflix Prize competition is launched by Netflix. The aim of the competition was to use machine learning to beat Netflix's own recommendation software's accuracy in predicting a user's rating for a film given their ratings for previous films by at least 10%. The prize was won in 2009.

Similarly in 2010, a website Kaggle was launched to serve as a platform for machine learning competitions.

These events are not some mind boggling breakthroughs or inventions but significant because it underscores the increased participation in machine learning from non academic people.
